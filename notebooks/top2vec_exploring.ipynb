## Explore a top2vec model


```python
import ndjson
import glob
import numpy as np
import pandas as pd
import top2vec
from top2vec import Top2Vec 
```

### Load primitives and filter events longer than 50 characters

```python
with open('../../corpus/primitives_220503/primitives_annotated.ndjson') as f:
    data = ndjson.load(f)
```

```python
# filter events longer than 50 characters
data_filtered = []
for item in data:
    if len(item['text']) > 50:
        data_filtered.append(item)

# prepare data for top2vec, create two lists of texts and document_ids
corpus = []
document_ids = []

for key in data_filtered:
    text = key['text']
    id = key['id']
    document_ids.append(id)
    corpus.append(text)
```

### Load a top2vec model

```python
# load model
model = Top2Vec.load('../../models/top2vec/top2vecmodel_220504')
```

```python
model.get_num_topics(reduced=True)
```

### Explore topics
